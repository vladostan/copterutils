```python
import numpy as np
import nrrd
import matplotlib.pyplot as plt
from matplotlib import cm
import os

%matplotlib inline
```

```python
from os import listdir
from os.path import isfile, join

imgdir = 'images'
files = [f for f in listdir(imgdir) if isfile(join(imgdir, f))]
```

```python
i = 2
file = files[i].split('.jpg')[0]

img = plt.imread('images/' + file + '.jpg')
# filtered = plt.imread('images_filtered/' + file + '.jpg')
data = nrrd.read('labels/' + file + '.nrrd')[0]
```

```python
fig, axes = plt.subplots()

fig.set_size_inches(5,5)

axes.imshow(img)
# axes[1].imshow(filtered)

fig.tight_layout()
```

```python
fig, axes = plt.subplots()

fig.set_size_inches(10,10)

cax = axes.imshow(data[:,:,0].T, cmap=cm.coolwarm)
cbar = fig.colorbar(cax, ticks=np.unique(data))
```
