```python
import os
import random
import matplotlib.pylab as plt
from glob import glob
import numpy as np
```

```python
PATH = os.path.abspath('data')

SOURCE_IMAGES = os.path.join(PATH, "images")

images = glob(os.path.join(SOURCE_IMAGES, "*.jpg"))

SOURCE_LABELS = os.path.join(PATH, "labels")

labels = glob(os.path.join(SOURCE_LABELS, "*.png"))
```

```python
r = random.sample(images, 3)

plt.figure(figsize=(16,16))
plt.subplot(231)
plt.imshow(plt.imread(r[0]))

plt.subplot(232)
plt.imshow(plt.imread(r[1]))

plt.subplot(233)
plt.imshow(plt.imread(r[2]))

for i in enumerate(r):
    r[i[0]] = r[i[0]].replace("images","labels")
    r[i[0]] = r[i[0]].replace("jpg","png")

plt.subplot(234)
plt.imshow(plt.imread(r[0]))

plt.subplot(235)
plt.imshow(plt.imread(r[1]))

plt.subplot(236)
plt.imshow(plt.imread(r[2]))

plt.tight_layout()
```

```python
x = [] # images
y = [] # labels

for img in images:
    full_size_image = plt.imread(img)
    x.append(full_size_image)

for lbl in labels:
    full_size_label = plt.imread(lbl,0)
    y.append(full_size_label)
```
