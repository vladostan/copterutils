```python
hex_colors = ['#000000','#4A4A4A','#FF0000', '#407700', '#00FF37', '#8B572A', '#0000FF', '#FF7600', '#50E3C2']
```

```python
rgb_colors = []

for hex_color in hex_colors:
    rgb_colors.append(webcolors.hex_to_rgb(hex_color))
```

```python
def colorize(mask, colors):
    colorMask = np.zeros([mask.shape[0], mask.shape[1], 3],dtype=np.uint8)
    for r in range(mask.shape[0]):
        for c in range(mask.shape[1]):
            colorMask[r,c,] = colors[mask[r,c]]

    return colorMask
```

```python
r = np.random.randint(80, size=3)

plt.figure(figsize=(16,16))
plt.subplot(231)
plt.imshow(x[r[0]])

plt.subplot(232)
plt.imshow(x[r[1]])

plt.subplot(233)
plt.imshow(x[r[2]])

plt.subplot(234)
plt.imshow(colorize(y[r[0]], colors = np.array(rgb_colors)))

plt.subplot(235)
plt.imshow(colorize(y[r[1]], colors = np.array(rgb_colors)))

plt.subplot(236)
plt.imshow(colorize(y[r[2]], colors = np.array(rgb_colors)))

plt.tight_layout()
```
