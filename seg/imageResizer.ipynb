```python
import PIL.Image as PIL
import os,re
```

```python
def get_new_path(path):
    basedir = os.path.dirname(path) + '/resized/'
    if not os.path.isdir(basedir):
        os.makedirs(basedir)
    base, ext = os.path.splitext(os.path.basename(path))

    return basedir + base + ext
```

```python
JPG = re.compile(".*\.(jpg|jpeg)", re.IGNORECASE)
PNG = re.compile(".*\.(png)", re.IGNORECASE)

# Images directory
images_dir = "/home/kenny/Desktop/copterutils/seg/data/images"

images = os.listdir(images_dir)

# Labels directory
labels_dir = "/home/kenny/Desktop/copterutils/seg/data/labels"

labels = os.listdir(labels_dir)
```

```python
images.sort()
labels.sort()
```

```python
n = 2

for image in images:
    if JPG.match(image):
        f = images_dir.rstrip("/") + "/" + image
        img = PIL.open(f)
        img = img.resize((img.size[0]//n, img.size[1]//n), PIL.NEAREST)
        img.save(get_new_path(f))
        print(get_new_path(f))

for label in labels:
    if PNG.match(label):
        f = labels_dir.rstrip("/") + "/" + label
        lbl = PIL.open(f)
        lbl = lbl.resize((lbl.size[0]//n, lbl.size[1]//n), PIL.NEAREST)
        lbl.save(get_new_path(f))
        print(get_new_path(f))
```
