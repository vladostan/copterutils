# Labeling using Supervise.ly annotation tool

The labeling process is done in [Supervise.ly](https://supervise.ly/) platform

![](http://cordelia.university.innopolis.ru/kamaz/winter_track/raw/Aerial_images_processing/labeling/example.png)

## Instructions

### Login

There are several accounts available for labeling:

> kamazbot1, kamazbot3, kamazbot4 (logins and passwords match)

> These bots have only "Annotate" permissions

**NOTE:** If you want own account with more permissions, ask [@vostankovich](https://t.me/vostankovich) to create one for you

***
    
### Project

    - Open the desired project out of two available (e.g. "Landscape segmentation from copter for Kamaz") 
    - Open the desired dataset inside the selected project (e.g. "copter")

**IMPORTANT**: If you want to add new images, you have to create new project, new dataset and redefine all classes again ðŸ˜Ÿ

***

### Labeling

![](http://cordelia.university.innopolis.ru/kamaz/winter_track/raw/Aerial_images_processing/labeling/menu.png?raw=true)

- Select the image you want to label

- Select "Add polygon" in left-side toolbar

- Select the class you want to label in the upper-left corner drop-menu

- There are 8 classes currently available for labeling (but can be extended):

    - *asphalt*
    - *building*
    - *forest*
    - *grass*
    - *ground*
    - *roadAsphalt*
    - *roadGround*
    - *water*

- Select the points for polygon
    
[Description of classes](https://new.supervise.ly/projects/198/classes/) is also available on Supervise.ly

**IMPORTANT**: All classes need to labeled using polygons

***

#### Tips

- Start labeling with large classes, end with smaller ones
- You can change the priority of classes in the right side context menu. If classes' areas intersect you can assign any class a higher priority.

![](http://cordelia.university.innopolis.ru/kamaz/winter_track/raw/Aerial_images_processing/labeling/classPriority.png)

##### Navigation 

- Move: Right mouse button
- Zoom: Scroll
- Change opacity: SHIFT + Scroll
- Finish labeling: Space
