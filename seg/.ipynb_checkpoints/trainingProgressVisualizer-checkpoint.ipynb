```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('logs/2018-11-06 19-32-27.log')
data.head()
```

```python
data[['acc','loss','val_acc', 'val_loss']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,10)
sns.lineplot(ax=ax, data = data[['acc','loss','val_acc', 'val_loss']], linewidth=2.5)
ax.set_xlabel('Epoch')
ax.set_title('Training progress')
fig.savefig('results/final/training_progress_2datasets.png')
```
