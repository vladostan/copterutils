```python
import matplotlib.pylab as plt
import numpy as np
```

```python
data = np.load("data.npz")
```

```python
x = data['x']
y = data['y']
```

```python
n_images, h, w = y.shape
```

```python
plt.imshow(y[0],cmap='gray')
```

```python
y = y.reshape([n_images, h*w])
y.shape
```

```python
cs = ['background','asphalt', 'building', 'forest', 'grass', 'ground', 'roadAsphalt', 'roadGround', 'water']
n_classes = len(cs)
temp = np.zeros(np.append(y.shape,n_classes),dtype='bool_')
temp.shape
```

```python
for i in range (n_images):
    for j in range(h*w):
            temp[i,j,(y[i,j])] = True
    print(i)
```

```python
y = temp.reshape([n_images,h,w,n_classes])
y.shape
del(temp)
```

```python
nrows, ncols = 3,3
fig, axes = plt.subplots(nrows=3, ncols=3)
fig.set_size_inches(15,10)
cl = 0
im = 59
for i in range(nrows):
    for j in range(ncols):
        axes[i,j].imshow(y[im,:,:,cl], cmap='gray')
        axes[i,j].set_title(cs[cl])
        cl += 1
        
fig.tight_layout()
```

```python
temp = np.zeros((n_images, h*w), dtype='uint8')
temp.shape
```

```python
y = y.reshape([n_images, h*w, n_classes])
y.shape
```

```python
for i in range (n_images):
    for j in range(h*w):
            temp[i,j] = np.argmax(y[i,j])
    print(i)
```

```python
y = temp
del(temp)
```

```python
y = y.reshape([n_images, h, w])
y.shape
```
